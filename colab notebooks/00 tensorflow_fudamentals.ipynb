```python
#create rank2 ie,dimension-2
import tensorflow as tf

rank_2_tensor = tf.constant([[10,7],
                           [3,4]])
rank_2_tensor.shape,rank_2_tensor.ndim
```

```python
#get the last ite of each row

rank_2_tensor[:,-1]
```

```python
#add extra dimension

rank_3_tensor = rank_2_tensor[...,tf.newaxis] #...=on all previous axis before last one
rank_3_tensor
```

```python
#or
rank_3_tensor = rank_2_tensor[:,:,tf.newaxis]
rank_3_tensor
```

```python
#alternative to tf.newaxis

tf.expand_dims(rank_2_tensor,axis=-1)#-1 means expand the final axis
```

```python
tf.expand_dims(rank_2_tensor,axis=1)#expand 0 axis
```

one of the most common

```python
tensor*tensor
tensor
```

```python
#using tensor flow

tf.matmul(tensor,tensor)
```

```python
#with python operator "@"
tensor@tensor
```

```python
x=tf.constant([[1,2],
               [3,4],
               [5,6]])

y=tf.constant([[7,8],
               [9,10],
               [11,12]])


x,y
```

```python
#matrix multiply of same shape
tf.matmul(x,y)
```

```python
#rules inner dimensions muust match
# result will have dimensions equal to outer dimensions
```

```python
#reshape
tf.reshape(y,shape=(2,3))
```

```python
#try to multiply now
x@tf.reshape(y,shape=(2,3))
```

```python
tf.matmul(tf.reshape(x,shape=(2,3)),y)
```

```python
#you can also use transpose
x,tf.transpose(x),tf.reshape(x,shape=(2,3))
```

```python
tf.matmul(tf.transpose(x),y)
```

### TENSOR AGGREGATION
defn: condensing them from multiple values to a smaller amount of values

max min mean  sum etc

```python
#create a random tensor
import numpy as np

a= tf.constant(np.random.randint(0,100,size=50))
a
```

```python
tf.size(a),a.ndim,a.shape
```

```python
#find the minimum
tf.reduce_min(a)
```

```python
#find the maximum
tf.reduce_max(a)
```

```python
#find the mean
tf.reduce_mean(a)
```

```python
#find the sum
tf.reduce_sum(a)
```

```python
import math
mean=tf.reduce_mean(a)
meanred=a-mean
meansq=meanred*meanred
sum=tf.reduce_sum(meansq)
variance=sum/50
standdev=math.sqrt(variance)

mean,meanred,meansq,sum,variance,standdev


```

```python
#OR
#variance=

import tensorflow_probability as tfp
tfp.stats.variance(a)
```

```python
#standard deviation =

tf.math.reduce_std(tf.cast(a,dtype=tf.float32))
```
